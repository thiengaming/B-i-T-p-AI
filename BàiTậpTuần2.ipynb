```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
quang_duong = ctrl.Antecedent(np.arange(0, 51, 1),  'quang_duong')
luu_luong   = ctrl.Antecedent(np.arange(0, 101, 1), 'luu_luong')
nhu_cau     = ctrl.Antecedent(np.arange(0, 101, 1), 'nhu_cau')
thoi_tiet   = ctrl.Antecedent(np.arange(0, 11, 1),  'thoi_tiet')
danh_gia    = ctrl.Antecedent(np.arange(1.0, 5.1, 0.1), 'danh_gia')
dung_gio    = ctrl.Antecedent(np.arange(0, 101, 1), 'dung_gio')
gia_cuoc    = ctrl.Consequent(np.arange(0, 101, 1), 'gia_cuoc')
diem_thuong = ctrl.Consequent(np.arange(0, 101, 1), 'diem_thuong')
quang_duong['ngan']       = fuzz.trimf(quang_duong.universe, [0, 0, 3])
quang_duong['trung_binh'] = fuzz.trimf(quang_duong.universe, [2, 5, 8])
quang_duong['dai']        = fuzz.trimf(quang_duong.universe, [6, 13, 20])
quang_duong['rat_xa']     = fuzz.trimf(quang_duong.universe, [15, 50, 50])
luu_luong['thap']       = fuzz.trimf(luu_luong.universe, [0, 0, 30])
luu_luong['trung_binh'] = fuzz.trimf(luu_luong.universe, [20, 45, 70])
luu_luong['cao']        = fuzz.trimf(luu_luong.universe, [60, 100, 100])
nhu_cau['thap']       = fuzz.trimf(nhu_cau.universe, [0, 0, 30])
nhu_cau['trung_binh'] = fuzz.trimf(nhu_cau.universe, [20, 45, 70])
nhu_cau['cao']        = fuzz.trimf(nhu_cau.universe, [60, 100, 100])
thoi_tiet['tot']        = fuzz.trimf(thoi_tiet.universe, [0, 0, 4])
thoi_tiet['trung_binh'] = fuzz.trimf(thoi_tiet.universe, [3, 5, 7])
thoi_tiet['xau']        = fuzz.trimf(thoi_tiet.universe, [6, 10, 10])
danh_gia['kem']        = fuzz.trimf(danh_gia.universe, [1.0, 1.0, 2.5])
danh_gia['trung_binh'] = fuzz.trimf(danh_gia.universe, [2.0, 3.0, 4.0])
danh_gia['tot']        = fuzz.trimf(danh_gia.universe, [3.5, 5.0, 5.0])
dung_gio['tre']      = fuzz.trimf(dung_gio.universe, [0, 0, 50])
dung_gio['dung_gio'] = fuzz.trimf(dung_gio.universe, [40, 60, 80])
dung_gio['som']      = fuzz.trimf(dung_gio.universe, [70, 100, 100])
gia_cuoc['thap']       = fuzz.trimf(gia_cuoc.universe, [0, 0, 30])
gia_cuoc['trung_binh'] = fuzz.trimf(gia_cuoc.universe, [20, 50, 70])
gia_cuoc['cao']        = fuzz.trimf(gia_cuoc.universe, [60, 80, 95])
gia_cuoc['rat_cao']    = fuzz.trimf(gia_cuoc.universe, [85, 100, 100])
diem_thuong['khong_co']   = fuzz.trimf(diem_thuong.universe, [0, 0, 25])
diem_thuong['it']         = fuzz.trimf(diem_thuong.universe, [15, 35, 55])
diem_thuong['trung_binh'] = fuzz.trimf(diem_thuong.universe, [45, 65, 80])
diem_thuong['cao']        = fuzz.trimf(diem_thuong.universe, [70, 100, 100])
r1  = ctrl.Rule(quang_duong['ngan'] & luu_luong['thap'] & nhu_cau['thap'], gia_cuoc['thap'])
r2  = ctrl.Rule(quang_duong['ngan'] & luu_luong['trung_binh'] & nhu_cau['cao'], gia_cuoc['trung_binh'])
r3  = ctrl.Rule(quang_duong['trung_binh'] & luu_luong['cao'] & nhu_cau['cao'], gia_cuoc['cao'])
r4  = ctrl.Rule(quang_duong['dai'] & luu_luong['trung_binh'] & thoi_tiet['tot'], gia_cuoc['trung_binh'])
r5  = ctrl.Rule(quang_duong['dai'] & luu_luong['cao'] & thoi_tiet['xau'], gia_cuoc['rat_cao'])
r6  = ctrl.Rule(quang_duong['rat_xa'] & luu_luong['cao'] & nhu_cau['cao'], gia_cuoc['rat_cao'])
r7  = ctrl.Rule(quang_duong['trung_binh'] & luu_luong['thap'] & nhu_cau['thap'], gia_cuoc['trung_binh'])
r8  = ctrl.Rule(quang_duong['ngan'] & luu_luong['cao'] & thoi_tiet['xau'], gia_cuoc['cao'])
r9  = ctrl.Rule(quang_duong['rat_xa'] & thoi_tiet['xau'], gia_cuoc['rat_cao'])
r10 = ctrl.Rule(quang_duong['trung_binh'] & luu_luong['trung_binh'] & thoi_tiet['trung_binh'], gia_cuoc['trung_binh'])
r11 = ctrl.Rule(danh_gia['tot'] & dung_gio['som'], diem_thuong['cao'])
r12 = ctrl.Rule(danh_gia['trung_binh'] & dung_gio['dung_gio'], diem_thuong['trung_binh'])
r13 = ctrl.Rule(danh_gia['kem'] & dung_gio['tre'], diem_thuong['khong_co'])
r14 = ctrl.Rule(quang_duong['dai'] & luu_luong['cao'] & dung_gio['dung_gio'], diem_thuong['cao'])
r15 = ctrl.Rule(quang_duong['trung_binh'] & luu_luong['trung_binh'] & danh_gia['tot'], diem_thuong['trung_binh'])
r16 = ctrl.Rule(danh_gia['kem'] & dung_gio['tre'], diem_thuong['khong_co'])
r17 = ctrl.Rule(quang_duong['rat_xa'] & thoi_tiet['xau'] & danh_gia['tot'], diem_thuong['cao'])
r18 = ctrl.Rule(quang_duong['ngan'] & danh_gia['trung_binh'] & dung_gio['dung_gio'], diem_thuong['it'])
r19 = ctrl.Rule(quang_duong['dai'] & luu_luong['cao'] & dung_gio['tre'], diem_thuong['it'])
r20 = ctrl.Rule(quang_duong['trung_binh'] & thoi_tiet['trung_binh'] & danh_gia['tot'], diem_thuong['trung_binh'])

gia_cuoc_ctrl = ctrl.ControlSystem([r1, r2, r3, r4, r5, r6, r7, r8, r9, r10])
diem_thuong_ctrl = ctrl.ControlSystem([r11, r12, r13, r14, r15, r16, r17, r18, r19, r20])
qd = 10
ll = 80
nc = 80
tt = 8
dg = 4.5
dgio = 85
sim_gia = ctrl.ControlSystemSimulation(gia_cuoc_ctrl)
sim_gia.input['quang_duong'] = qd
sim_gia.input['luu_luong'] = ll
sim_gia.input['nhu_cau'] = nc
sim_gia.input['thoi_tiet'] = tt
sim_gia.compute()
gia_out = sim_gia.output['gia_cuoc']

sim_diem = ctrl.ControlSystemSimulation(diem_thuong_ctrl)
sim_diem.input['quang_duong'] = qd
sim_diem.input['luu_luong'] = ll
sim_diem.input['thoi_tiet'] = tt
sim_diem.input['danh_gia'] = dg
sim_diem.input['dung_gio'] = dgio
sim_diem.compute()
diem_out = sim_diem.output['diem_thuong']

print(f"Giá cước (0-100): {gia_out:.2f}")
print(f"Điểm thưởng (0-100): {diem_out:.2f}")


gia_cuoc.view(sim=sim_gia)
diem_thuong.view(sim=sim_diem)
plt.show()
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

danh_gia = ctrl.Antecedent(np.arange(0, 5.1, 0.1), 'danh_gia')
khoi_luong = ctrl.Antecedent(np.arange(0, 1001, 1), 'khoi_luong')
bien_loi_nhuan = ctrl.Antecedent(np.arange(0, 101, 1), 'bien_loi_nhuan')
su_kien = ctrl.Antecedent(np.arange(0, 11, 1), 'su_kien')
doi_thu = ctrl.Antecedent(np.arange(0, 11, 1), 'doi_thu')
chiet_khau = ctrl.Consequent(np.arange(0, 71, 1), 'chiet_khau')

danh_gia['thap'] = fuzz.trimf(danh_gia.universe, [0, 0, 4.0])
danh_gia['trung_binh'] = fuzz.trimf(danh_gia.universe, [3.5, 4.25, 5.0])
danh_gia['cao'] = fuzz.trimf(danh_gia.universe, [4.5, 5.0, 5.0])

khoi_luong['thap'] = fuzz.trimf(khoi_luong.universe, [0, 0, 400])
khoi_luong['trung_binh'] = fuzz.trimf(khoi_luong.universe, [200, 500, 800])
khoi_luong['cao'] = fuzz.trimf(khoi_luong.universe, [600, 1000, 1000])

bien_loi_nhuan['thap'] = fuzz.trimf(bien_loi_nhuan.universe, [0, 0, 30])
bien_loi_nhuan['trung_binh'] = fuzz.trimf(bien_loi_nhuan.universe, [20, 50, 80])
bien_loi_nhuan['cao'] = fuzz.trimf(bien_loi_nhuan.universe, [60, 100, 100])

su_kien['khong_co'] = fuzz.trimf(su_kien.universe, [0, 0, 4])
su_kien['trung_binh'] = fuzz.trimf(su_kien.universe, [2, 5, 8])
su_kien['cao'] = fuzz.trimf(su_kien.universe, [6, 10, 10])

doi_thu['thap'] = fuzz.trimf(doi_thu.universe, [0, 0, 4])
doi_thu['trung_binh'] = fuzz.trimf(doi_thu.universe, [2, 5, 8])
doi_thu['cao'] = fuzz.trimf(doi_thu.universe, [6, 10, 10])

chiet_khau['rat_thap'] = fuzz.trimf(chiet_khau.universe, [0, 0, 5])
chiet_khau['thap'] = fuzz.trimf(chiet_khau.universe, [0, 7.5, 15])
chiet_khau['trung_binh'] = fuzz.trimf(chiet_khau.universe, [10, 15, 25])
chiet_khau['cao'] = fuzz.trimf(chiet_khau.universe, [15, 30, 45])
chiet_khau['rat_cao'] = fuzz.trimf(chiet_khau.universe, [35, 70, 70])

r1 = ctrl.Rule(danh_gia['cao'] & khoi_luong['cao'] & bien_loi_nhuan['cao'], chiet_khau['rat_thap'])
r2 = ctrl.Rule(danh_gia['thap'] & khoi_luong['thap'] & bien_loi_nhuan['cao'], chiet_khau['cao'])
r3 = ctrl.Rule(su_kien['cao'] & doi_thu['cao'], chiet_khau['rat_cao'])
r4 = ctrl.Rule(danh_gia['trung_binh'] & khoi_luong['trung_binh'] & bien_loi_nhuan['trung_binh'], chiet_khau['trung_binh'])
r5 = ctrl.Rule(doi_thu['thap'] & bien_loi_nhuan['thap'] & khoi_luong['cao'], chiet_khau['rat_thap'])
r6 = ctrl.Rule(danh_gia['thap'] & su_kien['khong_co'], chiet_khau['trung_binh'])
r7 = ctrl.Rule(khoi_luong['thap'] & bien_loi_nhuan['thap'], chiet_khau['rat_cao'])

he_thong_ctrl = ctrl.ControlSystem([r1, r2, r3, r4, r5, r6, r7])
mo_phong = ctrl.ControlSystemSimulation(he_thong_ctrl)

mo_phong.input['danh_gia'] = 4.3
mo_phong.input['khoi_luong'] = 500
mo_phong.input['bien_loi_nhuan'] = 10
mo_phong.input['su_kien'] = 9
mo_phong.input['doi_thu'] = 8

mo_phong.compute()
print(f"Mức chiết khấu đề xuất của hệ thống mờ: {mo_phong.output['chiet_khau']:.2f}%")
chiet_khau.view(sim=mo_phong)
plt.show()
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
mat_do     = ctrl.Antecedent(np.arange(0, 101, 1), 'mat_do')
khan_cap   = ctrl.Antecedent(np.arange(0, 101, 1), 'khan_cap')
tai_trong  = ctrl.Antecedent(np.arange(0, 101, 1), 'tai_trong')
giao_thong = ctrl.Antecedent(np.arange(0, 101, 1), 'giao_thong')
loi_nhuan  = ctrl.Antecedent(np.arange(0, 101, 1), 'loi_nhuan')
ket_hop_don = ctrl.Consequent(np.arange(0, 11, 1), 'ket_hop_don')
uu_tien     = ctrl.Consequent(np.arange(0, 101, 1), 'uu_tien')
mat_do['thap']       = fuzz.trimf(mat_do.universe, [0, 0, 40])
mat_do['trung_binh'] = fuzz.trimf(mat_do.universe, [20, 50, 80])
mat_do['cao']        = fuzz.trimf(mat_do.universe, [60, 100, 100])
khan_cap['thap']       = fuzz.trimf(khan_cap.universe, [0, 0, 40])
khan_cap['trung_binh'] = fuzz.trimf(khan_cap.universe, [20, 50, 80])
khan_cap['cao']        = fuzz.trimf(khan_cap.universe, [60, 100, 100])
tai_trong['thap']       = fuzz.trimf(tai_trong.universe, [0, 0, 40])
tai_trong['trung_binh'] = fuzz.trimf(tai_trong.universe, [20, 50, 80])
tai_trong['cao']        = fuzz.trimf(tai_trong.universe, [60, 100, 100])
giao_thong['thap']       = fuzz.trimf(giao_thong.universe, [0, 0, 40])
giao_thong['trung_binh'] = fuzz.trimf(giao_thong.universe, [20, 50, 80])
giao_thong['cao']        = fuzz.trimf(giao_thong.universe, [60, 100, 100])
loi_nhuan['thap']       = fuzz.trimf(loi_nhuan.universe, [0, 0, 40])
loi_nhuan['trung_binh'] = fuzz.trimf(loi_nhuan.universe, [20, 50, 80])
loi_nhuan['cao']        = fuzz.trimf(loi_nhuan.universe, [60, 100, 100])
ket_hop_don['it']     = fuzz.trimf(ket_hop_don.universe, [0, 0, 5])
ket_hop_don['mot_so'] = fuzz.trimf(ket_hop_don.universe, [0, 5, 10])
ket_hop_don['nhieu']  = fuzz.trimf(ket_hop_don.universe, [5, 10, 10])
uu_tien['thap']       = fuzz.trimf(uu_tien.universe, [0, 0, 50])
uu_tien['trung_binh'] = fuzz.trimf(uu_tien.universe, [0, 50, 100])
uu_tien['cao']        = fuzz.trimf(uu_tien.universe, [50, 100, 100])
r1 = ctrl.Rule(mat_do['cao'] & tai_trong['thap'] & giao_thong['thap'], ket_hop_don['nhieu'])
r2 = ctrl.Rule(mat_do['trung_binh'] & giao_thong['cao'] & khan_cap['trung_binh'], ket_hop_don['mot_so'])
r3 = ctrl.Rule(tai_trong['cao'] & mat_do['cao'] & loi_nhuan['trung_binh'], ket_hop_don['mot_so'])
r4 = ctrl.Rule(mat_do['thap'] & khan_cap['cao'] & giao_thong['trung_binh'], ket_hop_don['mot_so'])
r5 = ctrl.Rule(loi_nhuan['cao'] & khan_cap['cao'] & giao_thong['cao'], ket_hop_don['mot_so'])
r6 = ctrl.Rule(khan_cap['cao'] & loi_nhuan['cao'], uu_tien['cao'])
r7 = ctrl.Rule(khan_cap['trung_binh'] & giao_thong['trung_binh'], uu_tien['trung_binh'])
r8 = ctrl.Rule(khan_cap['thap'] & mat_do['cao'] & loi_nhuan['thap'], uu_tien['thap'])
r9 = ctrl.Rule(mat_do['cao'] & tai_trong['thap'] & giao_thong['trung_binh'], ket_hop_don['nhieu'])
ket_hop_ctrl = ctrl.ControlSystem([r1, r2, r3, r4, r5, r9])
uu_tien_ctrl = ctrl.ControlSystem([r6, r7, r8])
sim_ket_hop = ctrl.ControlSystemSimulation(ket_hop_ctrl)
sim_uu_tien = ctrl.ControlSystemSimulation(uu_tien_ctrl)
sim_ket_hop.input['mat_do']     = 80
sim_ket_hop.input['tai_trong']  = 20
sim_ket_hop.input['giao_thong'] = 50
sim_ket_hop.input['khan_cap']   = 50
sim_ket_hop.input['loi_nhuan']  = 50
sim_uu_tien.input['khan_cap']   = 50
sim_uu_tien.input['giao_thong'] = 50
sim_uu_tien.input['mat_do']     = 80
sim_uu_tien.input['loi_nhuan']  = 50
sim_ket_hop.compute()
sim_uu_tien.compute()
print(f" GỢI Ý KẾT HỢP ĐƠN HÀNG: {sim_ket_hop.output['ket_hop_don']:.0f} ")
print(f"MỨC ĐỘ ƯU TIÊN GIAO HÀNG: {sim_uu_tien.output['uu_tien']:.0f}% ")
ket_hop_don.view(sim=sim_ket_hop)
uu_tien.view(sim=sim_uu_tien)
plt.title("Đầu ra: Kết hợp đơn hàng")
plt.title("Đầu ra: Ưu tiên giao hàng")
plt.show()
```
