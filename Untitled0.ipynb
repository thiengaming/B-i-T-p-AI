```python
import folium
from geopy.geocoders import Nominatim
import osmnx as ox
import networkx as nx
address_A = "Ben Thanh Market, Ho Chi Minh City, Vietnam"
address_B = "Tan Son Nhat Airport, Ho Chi Minh City, Vietnam"
geolocator = Nominatim(user_agent="routing-example", timeout=10)
location_A = geolocator.geocode(address_A)
location_B = geolocator.geocode(address_B)
latlng_A = (location_A.latitude, location_A.longitude)
latlng_B = (location_B.latitude, location_B.longitude)
midpoint = ((latlng_A[0] + latlng_B[0]) / 2, (latlng_A[1] + latlng_B[1]) / 2)
G = ox.graph_from_point(midpoint, dist=3000, network_type='drive')
orig_node = ox.distance.nearest_nodes(G, latlng_A[1], latlng_A[0])
dest_node = ox.distance.nearest_nodes(G, latlng_B[1], latlng_B[0])
route = nx.shortest_path(G, orig_node, dest_node, weight='length')
route_coords = [(G.nodes[n]['y'], G.nodes[n]['x']) for n in route]
m = folium.Map(location=midpoint, zoom_start=14)
folium.Marker(latlng_A, popup="Điểm A", icon=folium.Icon(color='red')).add_to(m)
folium.Marker(latlng_B, popup="Điểm B", icon=folium.Icon(color='blue')).add_to(m)
folium.PolyLine(locations=route_coords, color='green', weight=5, opacity=0.8).add_to(m)
display(m)
```

```python
#bài 8
import folium
import osmnx as ox
import networkx as nx
a = [10.7705748, 106.6699228]
b = [10.775, 106.700]
midpoint = ((a[0] + b[0]) / 2, (a[1] + b[1]) / 2)
G = ox.graph_from_point(midpoint, dist=5000, network_type='drive')
drivers = [
    {'id': 'Thien', 'pos': [10.7289, 106.7082], 'type': 'bike'},
    {'id': 'Mai', 'pos': [10.7483, 106.6022], 'type': 'car'},
    {'id': 'Luong', 'pos': [10.7723,106.6983], 'type': 'car'}
]
customer_pos = a
customer_need = 'car'
best_driver = None
min_dist = float('inf')
final_route = None
cust_node = ox.distance.nearest_nodes(G, customer_pos[1], customer_pos[0])
for dr in drivers:
    if dr['type'] == customer_need:
        dr_node = ox.distance.nearest_nodes(G, dr['pos'][1], dr['pos'][0])
        try:
            dist = nx.shortest_path_length(G, dr_node, cust_node, weight='length')
            if dist < min_dist:
                min_dist = dist
                best_driver = dr
                final_route = nx.shortest_path(G, dr_node, cust_node, weight='length')
        except:
            continue
m = folium.Map(location=midpoint, zoom_start=15)
folium.Marker(customer_pos, popup="Khách tại UEH",icon=folium.Icon(color='red', icon='user')).add_to(m)
folium.Marker(b, popup="Nhà",icon=folium.Icon(color='blue', icon='home')).add_to(m)
for dr in drivers:
    icon_type = 'motorcycle' if dr['type'] == 'bike' else 'car'
    color = 'green' if dr == best_driver else 'orange'
    folium.Marker(dr['pos'], popup=f"{dr['id']} (Cách A: {min_dist:.0f}m nếu được chọn)",
                  icon=folium.Icon(color=color, icon=icon_type, prefix='fa')).add_to(m)
if final_route:
    route_coords = [(G.nodes[n]['y'], G.nodes[n]['x']) for n in final_route]
    folium.PolyLine(locations=route_coords, color='green', weight=6,opacity=0.8, tooltip="Đường tài xế đến đón").add_to(m)
m
```
